```python
import requests
from bs4 import BeautifulSoup
import json 
import codecs #File opening and closing support library.
```

```python
headers = {
    'user-agent': ('Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 '
                   '(KHTML, like Gecko) Chrome/48.0.2564.109 Safari/537.36')
}

PAGE = 1 # Current page number

result = []
```

```python
while True:
    response = requests.get(f"https://www.thegioididong.com/dtdd/iphone-11/danh-gia?p={PAGE}", headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")

    comments = soup.find_all("div", class_="comment__item par") #Find all div tags whose class name is comment__item par.
    if comments:
        print(f"Crawling data in page {PAGE}")
        data = []
        PAGE += 1
        for comment in comments:
            username = comment.find("div", class_="item-top").find("p", class_="txtname").text
            star = comment.find("div", class_="item-rate").find("div", class_="comment-star").find_all("i", class_="icon-star")
            content = comment.find("div", class_="comment-content").find("p", class_="cmt-txt").text

            data.append({"username": username, "content": content, "star": len(star)})

        result.append(data)
    else:
        break
```

```python
# Dump a dictionary to a JSON file.
with codecs.open("data.json", "w", encoding='utf-8') as outfile:
    json.dump(result, outfile, ensure_ascii=False)
```
